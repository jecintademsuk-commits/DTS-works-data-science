```python
def compute_average(scores):
    total = 0

    for score in scores:
        total = total + score

    average = total / len(scores)

    return average

scores = [45,66,76,88,99]

average = compute_average(scores)

print("Average score:", average)







```

```python
def check_pass_fail(score):
  if score >= 78:
     return "pass"
  else:
    return "fail"
print(check_pass_fail(66))
print(check_pass_fail(87))
```
