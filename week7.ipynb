```python
import matplotlib.pyplot as plt
```

```python
students = ['john','lusy','faith','joy','gath','mercy','johnn','samuel','david','caleb']
math_score = [99,88,98,78,67,58,56,99,89,78]
english_score = [88,98,78,67,9,88,90,78,76,56,]
science_score = [67,98,90,76,45,34,78,78,89,90]
```

```python
plt.figure()
plt.plot(students, math_score)
plt.title("math score - Line chart")
plt.xlabel("students")
plt.ylabel("score")
plt.show()
```

```python
plt.figure()
plt.bar(students, math_score)
plt.title("math score - Bar chart")
plt.xlabel("students")
plt.ylabel("score")
plt.show()
```

```python
plt.figure()
plt.pie(math_score,labels=students, autopct='%1.1f%%')
plt.title("math score Distribution")
plt.show
```
